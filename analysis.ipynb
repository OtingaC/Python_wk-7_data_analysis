```python
# Loading and exploring dataset
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris

# Load Iris dataset
iris = load_iris(as_frame=True)
df = iris.frame
print("Dataset successfully loaded")

# To display first few rows
df.head()
```

```python
# Explore structure
print("Dataset Info:")
print(df.info())

print("\nMissing Values:")
print(df.isnull().sum)
```

```python
# Clean Dataset
df = df.dropna()
print("Cleaned dataset (dropped missing values).")
```

```python
# Statistics
print("Basic Statistics")
df.describe()
```

```python
# Grouping - mean length by species
grouped = df.groupby("target")["petal length (cm)"].mean()
print("Average Petal Length per Species:\n", grouped)
```

```python
# Observation - Species with longest petals
longest_petal_species = grouped.idxmax()
print(f"Observation: Species {longest_petal_species} has longest average petals")
```

```python
# Line chart - sepal length over index
plt.figure(figsize=(8,5))
plt.plot(df.index, df["sepal length (cm)"], label="Sepal Length")
plt.title("Sepal Length Line Chart")
plt.xlabel("Index")
plt.ylabel("Sepal Length (cm)")
plt.legend()
plt.show()
```

```python
# Bar chart - average petal length by species
plt.figure(figsize=(8,5))
grouped.plot(kind="bar", color="skyblue")
plt.title("Average Petal Length per Species")
plt.xlabel("Species")
plt.ylabel("Average Petal Length (cm)")
plt.show()
```

```python
# Histogram - distribution of sepal width
plt.figure(figsize=(8,5))
plt.hist(df["sepal width (cm)"], bins=15, color="blue", edgecolor="black")
plt.title("Sepal Width Histogram")
plt.xlabel("Sepal Width (cm)")
plt.ylabel("Frequency")
plt.show()
```

```python
# Scatter plot - sepal length vs petal length
plt.figure(figsize=(8,5))
plt.scatter(df["sepal length (cm)"], df["petal length (cm)"], c=df["target"], cmap="viridis")
plt.title("Scatter Plot: Sepal Length vs Petal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.colorbar(label="Species")
plt.show()
```

```python
# Findings & Observations
print("\nEach species has distinct petal/sepal dimensions.")
print("Petal length is a strong distinguishing feature between species.")
print("Distributions show species overlap in sepal width, making it less discriminative.")
print("Scatter plots confirm separability of species via petal measurements.")
```
